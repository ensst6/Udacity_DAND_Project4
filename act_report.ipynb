# Summary of Analysis of "WeRateDogs" Twitter Data

## Question 1: Do more highly rated dogs receive more favorites and retweets?

Figure 1 compares favorite counts between dogs with high ratings (defined as >= 10) and those with low ratings (< 10). Higher-rated dogs average 9448 favorites, while lower-rated dogs average 2405. Comparing these values under the null hypothesis that favorite counts for higher-rated dogs are less than or equal two those of lower-rated dogs yields a p-value of 0.0 (no overlap of distributions) based on a simulated normal distribution.

Thus, dogs with a rating >= 10 receive significantly more favorites than dogs with a rating < 10.

Figure 2 compares retweet counts between dogs with high and low ratings. Higher-rated dogs average 2754 retweets, while lower-rated dogs average 817. Comparing these values under the null hypothesis that retweet counts for higher-rated dogs are less than or equal two those of lower-rated dogs yields a p-value of 0.0 (no overlap of distributions) based on a simulated normal distribution.

Thus, dogs with a rating >= 10 receive significantly more retweets than dogs with a rating < 10.

## Question 2: Do dog ratings differ by dog stage classification?

[WeRateDogs](https://twitter.com/dogrates) occasionally classifies dogs based on the stages described in [their book](https://www.amazon.com/WeRateDogs-Most-Hilarious-Adorable-Youve/dp/1510717145), i.e. "floofer", "pupper", "puppo", and "doggo". It is possible that the dogs which the rater takes the time to classify (of 2111 tweets in the dataset, 334 had staging classifications) are considered more interesting or special, and may therefore receive higher ratings.

Figure 3 compares ratings between dogs not given a stage classification ("None") and those in each of the four categories above.

A linear regression analysis on the relationship between stage and ratings shows that dogs classified as `doggo` and `puppo` have significantly higher ratings (p < 0.05) than those with no classification. 

Additionally, dogs classed as `pupper` have significantly lower ratings than `doggos` and `puppos`.

There were no significant differences between those classed as `doggo`, `floofer`, and `puppo`, nor between those classed as `floofer` and `pupper`.

While `floofers` appear to have higher ratings, the lack of statistical differences is likely due to the small number of dogs (n = 9) in this group.

It should also be noted that the R-squared value for this model is low (0.02), indicating that it does not explain much of the variation in ratings.

## Question 3: What are dogs most frequently mis-classified as by the neural network?

Neural network data in which images from the [WeRateDogs](https://twitter.com/dograte) were classified were also provided. These data are broken out by specific classification, prediction confidence, and whether or not the image was classified as a dog.

Figure 4 shows the top 10 objects that dogs were misclassified as based on the neural network's first prediction.

Dogs are most frequently mis-classified as seat belts, possibly due to presence of a collar or leash.

Other furry (teddy, hamster, tennis ball, siamese cat) things and a wild dog (dingo) are also common.

## Question 4: For dogs correctly classified as such, is the neural network's confidence related to the dog's rating?

This analysis examines the first prediction for dogs who were classified as dogs, irrespective of whether the breed classification was correct, since no data to validate the neural network's breed prediction were available.

Figure 5 shows the relationship between dog rating and the confidence of the neural network in its first prediction.

Linear regression yields a significant (p < 0.05) influence of rating on prediction confidence, with higher ratings associated with higher confidence. However, the R-squared value is 0.01, indicating the ratings do not expalain much of the variation in prediction confidence.
